```python
def newgender(uid):
    age,gender = uid
    if age < 22:
        return 'child'
    else:
        return gender
```

```python
#apply new gender definition
kdf['People']=kdf[['age','gender']].apply(newgender, axis=1)
```

```python
kdf.head()
```

```python
kdf['People'].value_counts()
```

```python
kdf1=DataFrame(kdf[kdf.age != 'Unknown'])
```

```python
kdf1['age'].value_counts()
```

```python
fig=sns.FacetGrid(kdf1,hue='gender',aspect=4)
fig.map(sns.kdeplot,'age',shade=True)
oldest=kdf1['age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```
